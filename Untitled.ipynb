```python
import json
import pandas as pd
import numpy as np
import requests
```

```python
file = 'business.json'
```

```python
with open(file, 'r', encoding='utf8', errors='ignore') as f:
    data = json.loads("[" +
        f.read().replace("}\n{", "},\n{") + "]")
```

```python
print(data[0])
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
df.shape
```

```python
mask = df['review_count'] < 100
df = df.loc[~mask, :]
df = df.reset_index(drop=True)
```

```python
df.shape
```

```python
df.head()
```

```python
train_state = input("Enter State to train Data: ")
train_city = input("Enter City to train Data: ")
```

```python
df['state'][2].lower() == train_state.lower()
df.shape
```

```python
df['state'] = df['state'].apply(lambda x: x.lower())
df['city'] = df['city'].apply(lambda x: x.lower())
```

```python
df['state'][2]
df['city'][2]
```

```python
mask = df['state'] != train_state.lower()
df = df.loc[~mask, :]
df = df.reset_index(drop=True)
mask = df['city'] != train_city.lower()
df = df.loc[~mask, :]
df = df.reset_index(drop=True)
```
